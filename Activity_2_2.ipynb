````python
import sys

#define what will be inputted in the code
def get_grade(label):
    grade = float(input(f"Enter Your {label} Grade: "))
    if grade > 100:
        print("Invalid Input")
        sys.exit()
    return grade
#return if the grade input exceeded the given range
def calculate_class_standing(HOA, QUI, ASSIGN):
    return HOA * 0.5 + QUI * 0.3 + ASSIGN * 0.2
#calculate the class standing
def calculate_grade(PCS, Exam):
    return PCS * 0.5 + Exam * 0.5
#calculate the grades
def summarize_grades(name, PG, MID, FIN):
    print("<------------------------------->")
    print("\nSUMMARY OF GRADES")
    print(f"Name: {name}")
    print(f"Prelim Grade: {PG}")
    print(f"Midterm Grade: {MID}")
    print(f"Final Grade: {FIN}")

# Input Name
name = input("Enter Your Name: ")
print("<------------------------------->\n")

# Prelim Grade
#input the prelim grades
print("PRELIM GRADE\n")
HOA = get_grade("Prelim Hands-on Activity")
QUI = get_grade("Prelim Quiz")
ASSIGN = get_grade("Prelim Assignment")
PCS = calculate_class_standing(HOA, QUI, ASSIGN)
print(f"Your Prelim Class Standing is: {PCS}")
PE = get_grade("Prelim Exam")
PG = calculate_grade(PCS, PE)
print(f"Your Prelim Grade is: {PG}\n")

# Midterm Grade
#input the midterm grades
print("MIDTERM GRADE\n")
MHOA = get_grade("Midterm Hands-on Activity")
MQUI = get_grade("Midterm Quiz")
MASSIGN = get_grade("Midterm Assignment")
MPCS = calculate_class_standing(MHOA, MQUI, MASSIGN)
print(f"Your Midterm Class Standing is: {MPCS}")
ME = get_grade("Midterm Exam")
MID = PG * 1/3 + (calculate_grade(MPCS, ME) * 2/3)
print(f"Your Midterm Grade is: {MID}\n")

# Final Grade
#input the final grades
print("FINAL GRADE\n")
FHOA = get_grade("Final Hands-on Activity")
FQUI = get_grade("Final Quiz")
FASSIGN = get_grade("Final Assignment")
FPCS = calculate_class_standing(FHOA, FQUI, FASSIGN)
print(f"Your Final Class Standing is: {FPCS}")
FE = get_grade("Final Exam")
FIN = MID * 1/3 + (calculate_grade(FPCS, FE) * 2/3)
print(f"Your Final Grade is: {FIN}\n")

# Summary of Grades
#print the sumnmarization of grades for better clarification
summarize_grades(name, PG, MID, FIN)
````
